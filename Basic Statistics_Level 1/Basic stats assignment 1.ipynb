Standardization

Standardization is a Critical aspect in Data Preprocessing.
This is the process of converting data into a uniform format ensuring its consistency.
This converts the Mean (average) to 0 & Standard Deviation (spread out of data from Mean) to 1.

One Hot Encoding

The process of converting the data (Categorical Data) into Binary language that Computers understands.
It converts all the unique values into seperate columns with values 0 (False) & 1 (True).
Each 1 value represents the value present in the Row for other columns & vice versa.
